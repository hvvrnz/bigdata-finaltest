```python
import pandas as pd
df = pd.DataFrame({'name':['kim', 'lee', 'park', 'choi'], 
                   'math':[70.0, 80, 90, 100],
                   'kor': [90, 70, 50, 95]})
df
```

```python
# 데이터 프레임도 하나의 객체이므로 df 내에 다양한 method 존재 
df.info()
```

```python
df['math']
```

```python
df[['name','kor']]
```

```python
df.name
```

```python
newdf = df[['name', 'kor', 'math']]
newdf
```

```python
df.iloc[0,0] # integer location iloc : 행번호와 열번호의 index를 적고 싶을 때
```

```python
df.iloc[:,[0,2,1]] # :만 기재하면 모든 행 전부 다 
```

```python
df.iloc[0:2,[0,2,1]] # 0:2...
```

```python
df
```

```python
df['합계'] = df.math + df.kor
df
```

```python
df.합계
```

```python
df['평균'] = df.합계 / 2
df
```

```python
df.sort_values(by = '평균', ascending = False) # default = ASC
```

```python
df.iloc[1,:] = ['LEE', 80, 80, 160, 80] # index 1행을 : (전체 열)
df
```

```python
df.sort_values(by = ['평균', 'math'], ascending = [True, False]) # 평균으로 먼저 정렬 후 그 다음 math로 정렬하라 
```
