```python
# 251117 서울 열린 데이터 광장 
# 주요 공원현황 API 실습
# 한 번에 1,000개까지 데이터 불러오기 가능

import requests

key = "53676b4a576a656e39326868534b7a"
url = f"http://openAPI.seoul.go.kr:8088/{key}/json/SearchParkInfoService/1/1000/"

r = requests.get(url)
content = r.json() # 리턴 결과가 Response [200]인 경우 메쏘드 .json() 이용
# content.keys() : 돌아오는 딕셔너리의 메인 키 이름이 서비스명과 같은 경우가 많다. 
results = content['SearchParkInfoService']
results.keys()
```

```python
# print(results['list_total_count'])
items = results['row']

for idx, item in enumerate(items):
    print(f"{idx + 1}")
    print(f"공원이름: {item['PARK_NM']}")
    print(f"주소: {item['PARK_ADDR']}")
    print('-' * 50)    
```

```python
# 서울시 지하철호선별 역별 승하차 인원 정보 API 실습

import requests

key = "53676b4a576a656e39326868534b7a"
url = f"http://openapi.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/1000/20251113/"

r = requests.get(url)
content = r.json()
# content.keys() : 돌아오는 딕셔너리의 메인 키 이름이 서비스명이 같음을 확인 
results = content['CardSubwayStatsNew']
results.keys()

results['list_total_count']
items = results['row']
items[0]

for item in items[:10]:
    print(f"호선: {item['SBWY_ROUT_LN_NM']}")
    print(f"역이름: {item['SBWY_STNS_NM']}")
    print(f"승차인원: {item['GTON_TNOPE']}")
    print(f"하차인원: {item['GTOFF_TNOPE']}")
    print("-" * 50)

```

```python
# 1호선만 추출 

for item in items:
    if item['SBWY_ROUT_LN_NM'] == "1호선":
        print(f"호선: {item['SBWY_ROUT_LN_NM']}")
        print(f"역이름: {item['SBWY_STNS_NM']}")
        print(f"승차인원: {item['GTON_TNOPE']}")
        print(f"하차인원: {item['GTOFF_TNOPE']}")
        print("-" * 50)
```

```python
import pandas as pd

line, stn, geton, getoff = [], [], [], []
for item in items:
    if item['SBWY_ROUT_LN_NM'] == "1호선":
        line.append(item['SBWY_ROUT_LN_NM'])
        stn.append(item['SBWY_STNS_NM'])
        geton.append(int(item['GTON_TNOPE']))
        getoff.append(int(item['GTOFF_TNOPE']))

df = pd.DataFrame({"호선": line, "역이름": stn, "승차인원": geton, "하차인원": getoff})
df
                   
```
