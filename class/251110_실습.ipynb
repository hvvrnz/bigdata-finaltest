```python
# 최종 code
# 에어코리아 측정소별 실시간 일평균 정보 조회 API 실습

import requests
import json

key = "eq85ap7zuwYvS0Q+prMd2sBzSEoYN9PmUKYjWMC/Q6KlksQkVEYWBwGvpXMIIdVGIq3+IZex0UsbUnohMvzqKQ==" #decoding
url = 'http://apis.data.go.kr/B552584/ArpltnStatsSvc/getMsrstnAcctoRDyrg'
params ={'serviceKey' : key, 'returnType' : 'json', 
         'numOfRows' : '1000', 'pageNo' : '1', 
         'inqBginDt' : '20251101', 'inqEndDt' : '20251101'}

# parameter 
# 네이버와 카카오는 maximum 100개만 가져올 수 있었는데, 여기는 1000개도 불러오기 가능 
#  'msrstnName' : '중앙탑면' 측정소를 지정하지 않으면 전국 측정소에 대한 정보가 다 나옴 따라서 하루 날짜로 지정함. 
# 'numOfRows' : '100'

response = requests.get(url, params=params)
# print(response.content)
results = json.loads(response.content) #json -> text 형식을 보기 좋게 딕셔너리로 변환
results = results['response'] # response key 안에 body(실제 내용) 와 header (정상여부 확인용)
body = results['body']
# body.keys()

print(f"검색 개수: {body['totalCount']}")
print("=" * 50)

items = body['items']

for idx, item in enumerate(items):
    print(f"{idx + 1}.")
    print(f"장소: {item['msrstnName']}")
    print(f"측정날짜: {item['msurDt']}")
    print(f"미세먼지: {item['pm10Value']}")
    print("-" * 50)
```
