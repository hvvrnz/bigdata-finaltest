```python
# 네이버 open API 검색 실습 - JSON

import requests
import urllib.request
import json
from bs4 import BeautifulSoup


client_id = "Ybd64TtAhG6ixrzVWC9k"
client_secret = "nMPmQ56nI1"

encText = urllib.parse.quote("충주 맛집")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText # JSON 결과
# blog.json json 표기 없으면 default json 
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # XML 결과

# 에러 안 난다는 가정 하에 예외처리 코드 삭제 
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)

# print(response_body.decode('utf-8')) # 결과 바로 출력시 디코딩 필요

    
response_body = response.read() # source code 
results = json.loads(response_body)
# 텍스트 형식의 json 응답형식을 딕셔너리 형태로 변환 or Soup으로 변환 
results

# items가 하나의 key

print(results['lastBuildDate'])
print()

items = results['items']
# 딕셔너리 변수와 key에 있고, 내용은 items 내에 있음.
for item in items:
    print(f"제목: {item['title']}")
    print(f"블로거 이름: {item['bloggername']}") #밖에서 "" 쓰면 안에서 '' 써야함.
    print(f"게시 날짜: {item['postdate']}")
    print("-" * 50)
```

```python
# 네이버 open API 검색 실습 - JSON

import requests
import urllib.request
import json
from bs4 import BeautifulSoup


client_id = "Ybd64TtAhG6ixrzVWC9k"
client_secret = "nMPmQ56nI1"

encText = urllib.parse.quote("파이썬")
url = "https://openapi.naver.com/v1/search/book.json?query=" 
url += encText
url += "&display=5&start=1&sort=sim"


# 에러 안 난다는 가정 하에 예외처리 코드 삭제 
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)


    
response_body = response.read() # source code 
results = json.loads(response_body)
# 텍스트 형식의 json 응답형식을 딕셔너리 형태로 변환 or Soup으로 변환 

# items가 하나의 key

print(results['lastBuildDate'])
print()

items = results['items']
# 딕셔너리 변수와 key에 있고, 내용은 items 내에 있음.
for item in items:
    print(f"제목: {item['title']}")
    print(f"저자: {item['author']}")
    # print(f"가격: {item['discount']:,}") -> 가격은 출력 형식을 보면 int형이 아니라 문자열 형태이므로 int 형으로 변환해야함. 
    print(f"가격: {int(item['discount']):,}")
    print(f"출판일: {item['pubdate']}")
    print("-" * 50)
```
