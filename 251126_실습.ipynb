```python
import pandas as pd

data = pd.read_excel("한국도로교통공단_일자별 시군구별 교통사고 통계_20241231.xlsx")
data
```

```python
jung = data[data.시군구 == "중구"]
jung
```

```python
data['시도시군구'] = data.시도 + " " + data.시군구
data
```

```python
newdata = data.iloc[:, [0,4,5,6,7,8,9]]
newdata
```

```python
# 그룹바이는 인덱스 여러개 지정 가능 (== SQL처럼..)
grp = newdata.groupby(["발생월", "시도시군구"])
result = grp.sum()
result = result.reset_index()

result[['시도', '시군구']] = result['시도시군구'].str.split(" ", expand=True)
result
```
