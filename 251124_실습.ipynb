```python
data = pd.read_csv("한국도로교통공단_일자별 시군구별 교통사고 통계_20241231.csv", encoding = "CP949")
data['부상자수'] = data.중상자수 + data.경상자수 + data.부상신고자수
data
```

```python
data.info()
```

```python
round(data.describe(),2)
```

```python
data[data.사고건수 >= 25]
```

```python
# 시도라는 변수로 그루핑하여 객체생성
grp = data.groupby("시도")
grp
```

```python
grp.size() # 개수
```

```python
grp[['사고건수', '부상자수']].sum()
```

```python
grp[['사고건수', '부상자수']].agg(['sum', 'max', 'mean'])
```

```python
cb = data[data.시도 == "충북"]
cb 
grp = cb.groupby("시군구")
grp[['사고건수', '부상자수']].agg(['sum', 'max', 'mean'])
```

```python
gg = data[data.시도 == "경기"]
gg 
grp = gg.groupby("시군구")
grp[['사고건수', '부상자수']].agg(['sum', 'max', 'mean'])
```

```python
# 각 시도, 월별로 사고건수에 대한 정보 나타내기 index = 행, columns = 열 values; 변수, func; 합
pd.pivot_table(data, index='시도', columns='발생월',
               values='사고건수', aggfunc='sum', margins=True)
```

```python
round(pd.pivot_table(data, index='시도', columns='발생월',
               values='부상자수', aggfunc='max', margins=True), 1)
```
