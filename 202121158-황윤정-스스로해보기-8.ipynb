```python
# 17 도전문제
# 서울에는 25개의 구가 있고, 2024년은 윤년으로 366일이다. 25개 구 중에 1년 동안 사고가 가장 많은 구의 사고 날은 365일이다. 
# 해당 구는 즉 1년 중 사고가 없는 날이 딱 1일인 구가 있다.
# 사고가 나지 않은 날이 1년에 1일뿐인 구를 모두 찾아 예시와같이 나타내시오.

seoul = df[df.시도 == '서울']
seoul
```

```python
seoul[seoul.사고건수 == 0] # 결과 없음 (사고가 난 날만 기록됨.)
```

```python
grp = seoul.groupby(['시군구'])
seoul1 = grp['사고건수'].agg('count').reset_index()
seoul1[seoul1.사고건수 == 365]
```

```python
# 표본 재생성
seoul2 = seoul[(seoul.시군구 == '강남구') |
                (seoul.시군구 == '송파구') | 
                (seoul.시군구 == '영등포구')]

seoul2
```

```python
grp1 = seoul2.groupby(['시군구', '발생월'])
seoul3 = grp1['발생일'].agg('sum').reset_index()

seoul3.columns = ['시군구', '발생월', '발생일합']
seoul3
```

```python
cond = ( ( ( seoul3['발생월'].isin([1,3,5,7,8,10,12]) ) & ( seoul3['발생일합'] != sum(range(1,32)) ) ) | # 31일까지 있는 month
        ( ( seoul3['발생월'].isin([2]) ) & ( seoul3['발생일합'] != sum(range(1,30)) ) ) | # 윤년 2월은 29일까지
        ( ( seoul3['발생월'].isin([4,6,9,11]) ) & ( seoul3['발생일합'] != sum(range(1,31)) ) ) ) # 30일까지 있는 month

seoul3[cond] # 강남구 3월, 송파구 7월, 영등포구 3월
```

```python
df2 = seoul3[cond].copy()
df2['사고 발생하지 않은 날'] = sum(range(1,32)) - df2.iloc[:,2]
df2
```

```python
seoul[ (seoul.시군구 == '강남구') & (seoul.발생월 == 3) & (seoul['발생일'] == 3) ]
seoul[ (seoul.시군구 == '송파구') & (seoul.발생월 == 7) & (seoul['발생일'] == 18) ]
seoul[ (seoul.시군구 == '영등포구') & (seoul.발생월 == 3) & (seoul['발생일'] == 2) ]
```

```python
gn = seoul[ (seoul.시군구 == '강남구') & (seoul.발생월 == 3) & (seoul['발생일'].isin([2,4])) ]
sp = seoul[ (seoul.시군구 == '송파구') & (seoul.발생월 == 7) & (seoul['발생일'].isin([17,19])) ]
ydp = seoul[ (seoul.시군구 == '영등포구') & (seoul.발생월 == 3) & (seoul['발생일'].isin([1,3])) ]

final = pd.concat([gn,sp,ydp], axis=0, ignore_index=True)
final
```

```python
# 17 최종 코드

seoul = df[df.시도 == '서울']

grp = seoul.groupby(['시군구'])
seoul1 = grp['사고건수'].agg('count').reset_index()
seoul1[seoul1.사고건수 == 365]

seoul2 = seoul[(seoul.시군구 == '강남구') |
                (seoul.시군구 == '송파구') | 
                (seoul.시군구 == '영등포구')]
grp1 = seoul2.groupby(['시군구', '발생월'])

seoul3 = grp1['발생일'].agg('sum').reset_index()
seoul3.columns = ['시군구', '발생월', '발생일합']
seoul3

cond = ( ( ( seoul3['발생월'].isin([1,3,5,7,8,10,12]) ) & ( seoul3['발생일합'] != sum(range(1,32)) ) ) |
        ( ( seoul3['발생월'].isin([2]) ) & ( seoul3['발생일합'] != sum(range(1,30)) ) ) |
        ( ( seoul3['발생월'].isin([4,6,9,11]) ) & ( seoul3['발생일합'] != sum(range(1,31)) ) ) )

seoul3[cond] 
df2 = seoul3[cond].copy()
df2['사고 발생하지 않은 날'] = sum(range(1,32)) - df2.iloc[:,2]
df2

gn = seoul[ (seoul.시군구 == '강남구') & (seoul.발생월 == 3) & (seoul['발생일'].isin([2,4])) ]
sp = seoul[ (seoul.시군구 == '송파구') & (seoul.발생월 == 7) & (seoul['발생일'].isin([17,19])) ]
ydp = seoul[ (seoul.시군구 == '영등포구') & (seoul.발생월 == 3) & (seoul['발생일'].isin([1,3])) ]

final = pd.concat([gn,sp,ydp], axis=0, ignore_index=True)
final

```
