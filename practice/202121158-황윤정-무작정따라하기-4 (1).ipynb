```python
# 웹 스크래핑
import requests
from bs4 import BeautifulSoup

url = "https://cyan.kku.ac.kr/~sunwoo/python/mypage.html"
r = requests.get(url)
r.encoding = "UTF-8"

html = r.text
soup = BeautifulSoup(html, "lxml")
soup
```

```python
# HTML 문서 제목 찾기

title_tag = soup.find("title")
title_tag
```

```python
# 제목 추출하기

title = title_tag.text
print(title)
```

```python
# h1 태그 검색하기 

h1_tag = soup.find("h1")
h1_tag
```

```python
# h1 제목 내용 추출하기

h1_tag.text
```

```python
# h2 태그 검색하기 - 1개

h2_tag = soup.find("h2")
print(h2_tag)

# 내용 추출하기
print(h2_tag.text)
```

```python
# h2 태그 모두 검색하기

h2_tags = soup.find_all("h2")
h2_tags
```

```python
# 내용 추출하기

for item in h2_tags:
    print(item.text)
```

```python
# 목록(li) 검색하기 / 내용 추출하기

li_tags = soup.find_all("li")
list = []
for item in li_tags:
    list.append(item.text)

list
```

```python
# 태그 + class로 검색하기
# 2025년 2학기 교과목만

li_2025_2_tags = soup.find_all("li", class_="a2025_2")
li_2025_2_tags
```

```python
# 내용 추출하기

for item in li_2025_2_tags:
    print(item.text)
```

```python
# 한국야구협회 - 기록실 (정규시즌 일자별 팀 순위)
import requests
from bs4 import BeautifulSoup

url = "https://www.koreabaseball.com/Record/TeamRank/TeamRankDaily.aspx"
r = requests.get(url)
html = r.text
soup = BeautifulSoup(html, "lxml")

# 기준 날짜 추출하기
date_tag = soup.find("span", class_="exp2")
date = date_tag.text
print(date + "\n")

# 테이블 헤드(제목) 추출하기
table_tag = soup.find("table", class_="tData")
thead = table_tag.find("thead")           # 테이블 헤드(head) 검색
th_tags = thead.find_all("th")            # 모든 th 셀 검색
head = []                                 # 테이블 head 추출하여 저장
for item in th_tags:
    head.append(item.text)

# 중간에 쉼표(,) 삽입하여 출력
print(", ".join(head))

# 테이블 내용 추출
tbody = table_tag.find("tbody")           # 테이블 body 검색
tr_tags = tbody.find_all("tr")            # 모든 행(tr) 검색
for tr in tr_tags:
    td_tags = tr.find_all("td")           # 각 행에서 모든 셀(td) 검색
    content = [x.text for x in td_tags]   # 각 셀(td) 내용 추출하여 저장
    print(", ".join(content))             # 쉼표(,) 삽입하여 출력
```
