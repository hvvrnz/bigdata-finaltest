```python
import pandas as pd

air = pd.read_excel("2025년7월-10월대기오염도.xlsx")
air
```

```python
air[['년', '월', '일']] = air['date'].str.split('-', expand=True)
air
```

```python
air.info()
```

```python
point = set(air['point'])
len(point)
```

```python
pm10na = air[air['pm10'].isna()]
grp = pm10na.groupby("시도")
grp.size()
```

```python
jn = air[air.시도 == "전남"]
grp = jn.groupby("시군구")
grp.size()
```

```python
cityName = ['광양시', '나주시', '목포시', '순천시', '여수시']
city = air[air['시군구'].isin(cityName)]
city
```

```python
city10 = city[city.월 == "10"]
city10
```

```python
grp = city10.groupby("시군구")
grp['pm10'].agg(['count', 'min','mean','max'])
```

```python
result = pd.pivot_table(city, index='시군구', columns='월', 
               values='pm25', aggfunc='mean', margins=True)
round(result, 2)
```
