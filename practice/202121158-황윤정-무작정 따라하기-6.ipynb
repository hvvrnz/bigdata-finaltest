```python
# 한국천문연구원 음양력 정보 - XML only

import requests
from bs4 import BeautifulSoup

key = "eq85ap7zuwYvS0Q+prMd2sBzSEoYN9PmUKYjWMC/Q6KlksQkVEYWBwGvpXMIIdVGIq3+IZex0UsbUnohMvzqKQ==" 
url = 'http://apis.data.go.kr/B090041/openapi/service/LrsrCldInfoService/getLunCalInfo'

solYear = 2024
solMonth = "%02d" % 11
solDay = "%02d" % 16
params ={'serviceKey' : key, 'solYear' : solYear, 
         'solMonth' : solMonth, 'solDay' : solDay}
response = requests.get(url, params=params)
soup = BeautifulSoup(response.content, "xml")


solWeek = soup.solWeek.text   # 요일 추출

lunYear = soup.find("lunYear").text
lunMonth = soup.lunMonth.text
lunDay = soup.lunDay.text

print(f"양력: {solYear}년 {solMonth}월 {solDay}일({solWeek}요일)")
print(f"음력: {lunYear}년 {lunMonth}월 {lunDay}일")
```
