```python
# 서울열린데이터광장 - 서울시 지하철 호선별 역별 승하차 인원 정보

import requests

key = "53676b4a576a656e39326868534b7a"
date = "20251101"
url = f"http://openAPI.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/999/{date}"
r = requests.get(url)

content = r.json()
results = content['CardSubwayStatsNew']

print(f"검색 개수: {results['list_total_count']}")
print()

items = results['row']
items[0]
```

```python
# 모든 결과 출력
for item in items:
    print(f"날짜: {item['USE_YMD']}")
    print(f"호선: {item['SBWY_ROUT_LN_NM']}")
    print(f"역이름: {item['SBWY_STNS_NM']}")
    print(f"승차인원: {item['GTON_TNOPE']}")
    print(f"하차인원: {item['GTOFF_TNOPE']}")
    print("-"*50)
```

```python
# 호선 정보만 출력
line = []

for item in items:
    line.append(item['SBWY_ROUT_LN_NM'])

print(f"전체 개수: {len(line)}")

# 리스트 line을 집합(set) lines로 저장
# 집합은 중복을 제거하고 1개씩만

lines = set(line)
print(f"호선 수: {len(lines)}")
lines
```

```python
# 1호선만

for item in items:
    if item['SBWY_ROUT_LN_NM'] == "1호선":
        print(f"{item['SBWY_STNS_NM']:<20}", end=" ")
        print(f"{item['GTON_TNOPE']:>10}", end=" ")
        print(f"{item['GTOFF_TNOPE']:>10}")
```

```python
# 데이터프레임으로 저장
import pandas as pd

stn, geton, getoff = [], [], []
for item in items:
    if item['SBWY_ROUT_LN_NM'] == "1호선":
        stn.append(item['SBWY_STNS_NM'])
        geton.append(int(item['GTON_TNOPE']))
        getoff.append(int(item['GTOFF_TNOPE']))

line_1 = pd.DataFrame({'stn': stn, 'geton': geton, 'getoff': getoff})
line_1
```

```python
# 데이터 프레임 정보 확인
line_1.info()
```

```python
# 각 변수의 기술통계량(숫자 변수만)
line_1.describe()
```

```python
# 그래프로 나타내기 - 역 순서 무시

import matplotlib.pyplot as plt
from matplotlib import rc

rc('font', family='NanumGothic')

#rc('font', family='AppleGothic')     # 맥북
#plt.rcParams['font.family'] = 'HCR Dotum'  # vscode 

plt.figure(figsize=(12,5))
plt.plot(range(1,11), line_1.geton, 'ro-')
plt.plot(range(1,11), line_1.getoff, 'bs-')
plt.xticks(range(1,11), line_1.stn)
plt.ylim(0,100000)
plt.show()
```

```python
# 서울열린데이터광장 - 서울시 도서관 이용시간 및 휴관일 정보 

import requests

key = "53676b4a576a656e39326868534b7a"
url = f"http://openapi.seoul.go.kr:8088/{key}/json/SeoulLibraryTimeInfo/1/10/"
r = requests.get(url)
content = r.json()
content.keys()
```

```python
# 총 검색 개수 및 결과 확인
results = content['SeoulLibraryTimeInfo']
items = results['row']

print(f"총 개수: {results['list_total_count']}")
print(f"결과 개수: {len(items)}") 

items[0]
```

```python
# 결과 추출.출력
for idx, item in enumerate(items):
    print(f"{idx+1}.")
    print(f"도서관명: {item['LBRRY_NAME']}")
    print(f"구: {item['CODE_VALUE']}")
    print(f"주소: {item['ADRES']}")
    print(f"휴관일: {item['FDRM_CLOSE_DATE']}")
    print("-"*50)
```

```python
# 서울열린데이터광장 - 도서관 모든 결과 호출
# 총 검색 결과 1532개

import requests

key = "53676b4a576a656e39326868534b7a"
NAME, ADDR, GU, CLOSE  = [], [], [], []

total = 1532
cnt_per_page = 1000

for start in range(1, total+1, cnt_per_page):
    end = start+999   # 1-1000, 1001-2000으로 지정하기 위함
    url = f"http://openapi.seoul.go.kr:8088/{key}/json/SeoulLibraryTimeInfo/{start}/{end}/"
    r = requests.get(url)
    content = r.json()
    results = content['SeoulLibraryTimeInfo']  
    items = results['row']
    
    for item in items:
        NAME.append(item['LBRRY_NAME'])
        GU.append(item['CODE_VALUE'])
        ADDR.append(item['ADRES'])
        CLOSE.append(item['FDRM_CLOSE_DATE'])        
len(NAME)
```

```python
# 빈도표 확인
from collections import Counter

freq = Counter(GU)
print(freq)
```

```python
# 빈도표를 데이터프레임으로 저장 
import pandas as pd

df = pd.DataFrame(freq.items(), columns=['GU', 'count'])
print(df)
```
