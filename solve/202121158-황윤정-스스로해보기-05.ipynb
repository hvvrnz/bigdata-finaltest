```python
# 6 <도전 문제 2>
# 카카오 Open API 카페 검색 (검색어: 충주맛집)

import requests
from bs4 import BeautifulSoup
import html

# HTML 태그, 엔티티 제거 함수
def clean_html_text(text):
    text = html.unescape(text)
    if "<" in text and ">" in text:
        # 문자열에 <, >가 포함된 경우 BeautifulSoup이 혼동할 수 있어 경고가 발생하기도 하여 조건 추가 
        text = BeautifulSoup(text, "html.parser").get_text()
    return text

# 기본 설정
url = "https://dapi.kakao.com/v2/search/cafe"
headers = {"Authorization": "KakaoAK c699f3fbd5bb2b3f9be278dd0fb3e227"}
params = {"query": "충주맛집", "sort": "accuracy", "page": 1, "size": 50}

# 결과 저장용 리스트
titles, contents, dates = [], [], []

res = requests.get(url, headers=headers, params=params)
results = res.json()
meta = results["meta"]

pageable = meta["pageable_count"]
size = params["size"]
pages = (pageable // size) + (1 if pageable % size > 0 else 0)

print(f"page={pages} {meta}")
print("=" * 50)



while True:
    res = requests.get(url, headers=headers, params=params)
    results = res.json()

    documents = results["documents"]
    meta = results["meta"]

    for doc in documents:
        titles.append(clean_html_text(doc["title"]))
        contents.append(clean_html_text(doc["contents"]))
        dates.append(doc["datetime"])

    if meta["is_end"]:
        break

    params["page"] += 1  # 다음 페이지로 이동

for i in range(3):
    print(f"{i+1}. 제목: {titles[i]}")
    print(f"   내용: {contents[i]}")
    print(f"   날짜: {dates[i]}")
    print("-" * 50)

for i in range(-3, 0):
    print(f"{len(titles)+i+1}. 제목: {titles[i]}")
    print(f"   내용: {contents[i]}")
    print(f"   날짜: {dates[i]}")
    print("-" * 50)
```
