```python
# 3

import requests
import pandas as pd
from bs4 import BeautifulSoup

url = "https://www.moe.go.kr/boardCnts/listRenew.do?boardID=333&m=020501&s=moe"
r = requests.get(url)
r.encoding = "UTF-8"
html = r.text
soup = BeautifulSoup(html, "html.parser")

tables = soup.find_all("table")
for table3 in tables:
    caption = table3.find("caption")
        # 하나라도 <caption>이 없다면, text.strip() 수행 못하고 예외발생 할 수 있으므로 예외처리
    if not caption:
        continue 
    caption_text = caption.text.strip()
    
    if "공지사항 목록" in caption_text:
        target_table = table3
        break
        # break을 꼭 적어줘야 함.
        # 모든 테이블에 <caption>이 있음	정상 동작하지만, 하나라도 <caption> 누락 -> 예외 발생 
        # break를 사용하면 조건에 맞는 첫 번째 테이블을 반환
        # 찾은 뒤에는 즉시 반복을 중단하여 불필요한 탐색을 방지할 수 있도록


# 헤더 추출
thead = target_table.find("thead")         
th_tags = thead.find_all("th")           
head = []                                 
    
for item in th_tags:
    head.append(item.text)
print(", ".join(head))


# 내용 추출
# 번호, 제목, 등록자, 등록일, 조회수 순 출력
# <tr></tr> 당 게시글 하나 

tbody = target_table.find("tbody") 
tr_tags = tbody.find_all("tr")       

# 번호, 제목
# 제목을 그냥 td로 추출하면 개행 공백 문자 전부 출력되어 <a> 태그 안에 title 내용만 출력하도록 코드 작성

rows = [] 

for tr in tr_tags:
    td = tr.find_all("td")
    td1 = tr.find("td", class_ = "no") 
    no = td1.text
    
    td2 = tr.find("td", class_="title left") 
    a_tag = td2.find("a")
    title = a_tag.get("title") # 제목
    # text = a_tag.text.strip() -> 개행문자 다 가져옴 

    # 등록자, 등록일, 조회수
    no_class_tds = [td for td in tr.find_all("td") if not td.has_attr("class")]
    content = [x.text for x in no_class_tds]   
    
    writer = content[0]
    date = content[1]
    view = content[2]

    print(f"{no}, {title}, {writer}, {date}, {view}")
    rows.append([no, title, writer, date, view])
       

    


# <table>
# <caption>공지사항 목록 : 번호, 제목, 등록자, 등록일, 조회수의 내용을 확인 하실 수 있습니다.</caption>
# <col style="width:5em"/>
# <col style="width:auto"/>
# <col style="width:7em"/>
# <col style="width:7em"/>
# <col style="width:5em"/>
# <thead>
# <tr>
# <th scope="col">번호</th>
# <th scope="col">제목</th>
# <th scope="col">등록자</th>
# <th scope="col">등록일</th>
# <th scope="col">조회수</th>
# </tr>
# </thead>
         
# <tbody>
# <tr>
# <td class="no">1741</td>
# <td class="title left">
# <a href="#" onclick="javascript:goView('333', '104244', '0', null, 'W', '1', 'N', '');" title="「2025년 대한민국 우수교육시설」작품 공모 안내(변경)">
# 					         				「2025년 대한민국 우수교육시설」작품 공모 안내(변경)
# 					         			</a>
# </td>
# <td>이윤범</td>
# <td>2025-10-30</td>
# <td>187</td>
# </tr>
# ... 
         
# <tr>
# <td class="no">1732</td>
# <td class="title left">
# <a href="#" onclick="javascript:goView('333', '103737', '0', null, 'W', '1', 'N', '');" title="2025 대한민국 인재상 선발 공고">
# 					         				2025 대한민국 인재상 선발 공고
# 					         			</a>
# </td>
# <td>장명훈</td>
# <td>2025-07-15</td>
# <td>6582</td>
# </tr>
# </tbody>
# </table>

```
