```python
# 01
# 서울열린데이터광장 - 지하철 8호선 역이름, 승차인원, 하차인원 출력

import requests

key = "53676b4a576a656e39326868534b7a"
url = f"http://openAPI.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/1000/20251118/"
r = requests.get(url)
content = r.json()
items = content['CardSubwayStatsNew']['row']

print(f"{'역이름':<5} {'승차인원':<5} {'하차인원':<5}")
for item in items:
    if item['SBWY_ROUT_LN_NM'] == "8호선":
        print(item['SBWY_STNS_NM'] + ",", end=" ")
        print(item['GTON_TNOPE'] + ",", end=" ")
        print(item['GTOFF_TNOPE'])
        print("-" * 50)
```

```python
# 02
import pandas as pd

stn, geton, getoff, total = [], [], [], []
for item in items:
    if item['SBWY_ROUT_LN_NM'] == "8호선":
        stn.append(item['SBWY_STNS_NM'])
        geton.append(int(item['GTON_TNOPE']))
        getoff.append(int(item['GTOFF_TNOPE']))
        total.append(int(item['GTON_TNOPE']) + int(item['GTOFF_TNOPE']))

df = pd.DataFrame({"역이름": stn, "승차인원": geton, "하차인원": getoff, "합계": total})
df
                  
```

```python
# 03
# 10월 1일부터 10월 31일까지 고속터미널역에 대하여 
#날짜, 호선, 역이름, 승차인원, 하차인원을 각각 리스트 변수로 저장후 
# pandas의 데이터프레임으로 저장하고 출력
import requests
import pandas as pd

key = "53676b4a576a656e39326868534b7a"
dates, line, stn, geton, getoff = [], [], [], [], []
for n in range(1,32):
    date = "202510" + f"{n:02d}"
    url = f"http://openAPI.seoul.go.kr:8088/{key}/json/CardSubwayStatsNew/1/1000/{date}/"
    r = requests.get(url)
    content = r.json()
    items = content['CardSubwayStatsNew']['row']

   
    for item in items:
        if item['SBWY_STNS_NM'] == "고속터미널":
            dates.append(int(item['USE_YMD']))
            line.append(item['SBWY_ROUT_LN_NM'])
            stn.append(item['SBWY_STNS_NM'])
            geton.append(item['GTON_TNOPE'])
            getoff.append(item['GTOFF_TNOPE'])
           
df = pd.DataFrame({"date": dates, "line": line, "stn": stn, "geton": geton, "getoff": getoff})
df
```

```python
# 04
# 서울열린데이터광장 - 서울시 버스정류소 위치정보

import requests

key = "53676b4a576a656e39326868534b7a"
url = f"http://openAPI.seoul.go.kr:8088/{key}/json/busStopLocationXyInfo/1/1000/"
r = requests.get(url)
content = r.json()
items = content['busStopLocationXyInfo']['row']

count = 0
for item in items:
    print("정류장번호: " + item['STOPS_NO'])
    print("정류장이름: " + item['STOPS_NM'])
    print(f"(위도, 경도): ({item['YCRD']}, {item['XCRD']})")
    print("-" * 50)

    count += 1
    if count == 10:
        break
```
