```python
# 4
# 한국천문연구원 음양력정보 서비스 중 음력일정보 조회
# XML only

import requests
from bs4 import BeautifulSoup

key = "eq85ap7zuwYvS0Q+prMd2sBzSEoYN9PmUKYjWMC/Q6KlksQkVEYWBwGvpXMIIdVGIq3+IZex0UsbUnohMvzqKQ==" 
url = 'http://apis.data.go.kr/B090041/openapi/service/LrsrCldInfoService/getLunCalInfo'

print(f"{'양력':<16}음력")
solYear = 2025
solMonth = '11'
for num in range(1,31):
    solDay = "%02d" % num
    params ={'serviceKey' : key, 'solYear' : solYear, 'solMonth' : solMonth, 'solDay': solDay}
    response = requests.get(url, params=params)
    soup = BeautifulSoup(response.content, "xml")

    solWeek = soup.solWeek.text   # 요일 추출

    lunYear = soup.find("lunYear").text
    lunMonth = soup.lunMonth.text
    lunDay = soup.lunDay.text

    print(f"{solYear}-{solMonth}-{solDay}({solWeek})", end=" -> ")
    print(f"{lunYear}-{lunMonth}-{lunDay}")
```
